```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense, LSTM
import os

plt.style.use('fivethirtyeight')

# Configuration
cryptos = ['BTC-USD', 'ETH-USD', 'DOGE-USD']  
years_of_data = 15
base_days = 100
future_days_to_predict = 10

def analyze_crypto(crypto):
    # Download data
    end = datetime.now()
    start = datetime(end.year - years_of_data, end.month, end.day)
    
    print(f"\nAnalyzing {crypto}...")
    print(f"Downloading data from {start} to {end}")
    data = yf.download(crypto, start=start, end=end)
    
    if data.empty:
        print(f"No data available for {crypto}")
        return
    
    # Process closing prices
    closing_price = data[['Close']].copy()
    closing_price.columns = ['Close']
    
    # Calculate moving averages
    closing_price['MA_365'] = closing_price['Close'].rolling(window=365).mean()
    closing_price['MA_100'] = closing_price['Close'].rolling(window=100).mean()

    # Plot closing prices with moving averages
    plt.figure(figsize=(15, 6))
    plt.plot(closing_price.index, closing_price['Close'], label='Close Price', linewidth=2)
    plt.plot(closing_price.index, closing_price['MA_365'], 'r--', label='365 Days MA', linewidth=2)
    plt.plot(closing_price.index, closing_price['MA_100'], 'g--', label='100 Days MA', linewidth=2)
    plt.title(f"{crypto} Price History with Moving Averages", fontsize=16)
    plt.xlabel("Date", fontsize=14)
    plt.ylabel('Price (USD)', fontsize=14)
    plt.legend(fontsize=12)
    plt.grid(alpha=0.3)
    plt.show()

    # Prepare data for LSTM
    scaler = MinMaxScaler(feature_range=(0, 1))
    scaled_data = scaler.fit_transform(closing_price[['Close']].dropna())

    # Create training data structure
    X, y = [], []
    for i in range(base_days, len(scaled_data)):
        X.append(scaled_data[i-base_days:i])
        y.append(scaled_data[i])
        
    X, y = np.array(X), np.array(y)
    
    # Split data
    train_size = int(len(X) * 0.9)
    X_train, X_test = X[:train_size], X[train_size:]
    y_train, y_test = y[:train_size], y[train_size:]

    # Build LSTM model
    model = Sequential([
        LSTM(128, return_sequences=True, input_shape=(X_train.shape[1], 1)),
        LSTM(64, return_sequences=False),
        Dense(25),
        Dense(1)
    ])

    model.compile(optimizer='adam', loss='mean_squared_error')
    
    # Train model
    print(f"\nTraining model for {crypto}...")
    model.fit(X_train, y_train, batch_size=5, epochs=10, verbose=1)

    # Make predictions
    predictions = model.predict(X_test)
    predictions = scaler.inverse_transform(predictions)
    y_test_actual = scaler.inverse_transform(y_test)

    # Plot predictions
    plot_data = pd.DataFrame({
        'Actual': y_test_actual.flatten(),
        'Predicted': predictions.flatten()
    }, index=closing_price.index[train_size + base_days:])

    plt.figure(figsize=(15, 6))
    plt.plot(plot_data.index, plot_data['Actual'], label='Actual Price', linewidth=2)
    plt.plot(plot_data.index, plot_data['Predicted'], label='Predicted Price', linewidth=2)
    plt.title(f"{crypto} Price Predictions vs Actual", fontsize=16)
    plt.xlabel("Date", fontsize=14)
    plt.ylabel('Price (USD)', fontsize=14)
    plt.legend(fontsize=12)
    plt.grid(alpha=0.3)
    plt.show()

    # Future predictions
    last_100_days = scaled_data[-base_days:].reshape(1, -1, 1)
    future_predictions = []
    
    print(f"\nPredicting next {future_days_to_predict} days for {crypto}...")
    for _ in range(future_days_to_predict):
        pred = model.predict(last_100_days)
        future_predictions.append(scaler.inverse_transform(pred)[0][0])
        last_100_days = np.append(last_100_days[:, 1:, :], pred.reshape(1, 1, -1), axis=1)

    # Plot future predictions
    plt.figure(figsize=(15, 6))
    plt.plot(range(future_days_to_predict), future_predictions, marker='o', 
             color='purple', linewidth=2, label='Predicted Prices')
    
    for i, price in enumerate(future_predictions):
        plt.text(i, price, f'{price:,.2f}', ha='center', va='bottom', fontsize=10)
        
    plt.title(f"{crypto} {future_days_to_predict}-Day Price Forecast", fontsize=16)
    plt.xlabel("Days in Future", fontsize=14)
    plt.ylabel('Predicted Price (USD)', fontsize=14)
    plt.xticks(range(future_days_to_predict), [f'Day {i+1}' for i in range(future_days_to_predict)])
    plt.legend(fontsize=12)
    plt.grid(alpha=0.3)
    plt.show()

    # Save model
    model_name = f"{crypto.replace('-', '_')}_model.keras"
    model.save(model_name)
    print(f"Model saved as {model_name}")

# Create output directory if it doesn't exist
os.makedirs('crypto_analysis', exist_ok=True)

# Analyze each cryptocurrency
for crypto in cryptos:
    try:
        analyze_crypto(crypto)
    except Exception as e:
        print(f"Error analyzing {crypto}: {str(e)}")
```
